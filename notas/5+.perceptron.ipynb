#### Funci贸n segmentada
$$
    f(x) =
    \begin{cases}
    0,&\quad\text{para } x<0\\
    1,&\quad\text{para } x\ge0\\
    \end{cases}
$$

```python
def H(x):
    y = np.zeros_like(x)
    for i, v in enumerate(x):
        if v < 0:
            y[i] = 0
        elif v >= 0:
            y[i] = 1
    return y

x = np.linspace(-10,10,1000) 
y = H(x)
plt.plot(x, y)
plt.grid()
```
