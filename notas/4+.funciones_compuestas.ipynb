### Manipulacin de funciones 
-----------------------------

+ **Desplazamientos horizontales y verticales**
        
    Siendo $c$ una constante mayor que cero, entoces la grfica:
    
    - $y = f(x) + c$ donde $c$ es un dezplazamiento hacia arriba. 
    - $y = f(x) - c$ donde $c$ es un dezplazamiento hacia abajo.
    - $y = f(x - c)$ donde $c$ es un dezplazamiento a la derecha. 
    - $y = f(x + c)$ donde $c$ es un dezplazamiento a la izquierda.

```python
## Experimentacin
def f(x): ## Funcin 
    return x**2

c = 5 ## Dezplazamiento
x = np.linspace(-5,5,num= 100)
y = f(x + c)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

+ **Alargamientos horizontales y verticales**
    
    Siendo $c$ una constante mayor que cero, entonces la grfica.
    
    - $y = f(x)  c$ alarga la grfica verticalmente en un factor de $c$.
    - $y = f(x)  \frac{1}{c}$ comprime la grfica verticalmente en un factor de $c$
    - $y = f(x  c)$ comprime la grfica horizontalmente en un factor de $c$
    - $y = f(x  \frac{1}{c})$ alarga la grfica horizontalmente en un factor de $c$

```python
## Experimentacin
def f(x): ## Funcin 
    return np.sin(x)

c = 5 ## Dezplazamiento
x = np.linspace(-10,10,num= 1000)
y = f(x/5)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```
